```python
import pandas as pd
import numpy as np
import geojson
```

```python
p_df = pd.read_excel('data/Pacific_Macbeth.xlsx')
p_df = p_df[p_df['Play'] == 'Macbeth']
p_df
```

```python
f_acc = []
for ix, r in p_df.iterrows():
    g = geojson.Point((r['Longitude'], r['Latitude']))
    f = geojson.Feature(geometry=g, properties= {'Theatre': r['Theatre'], 'City': r['City'], 'State': r['State'],
                                                'Day': r['Day'], 'Month': r['Month'], 'Year': r['Year']})
    f_acc.append(f)

f_col = geojson.FeatureCollection(f_acc)

with open('data/Performances.geojson', 'w') as f:
    geojson.dump(f_col, f)
```

```python
c_acc = []
for ix, r in p_df.iterrows():
    g = (r['Longitude'], r['Latitude'])
    c_acc.append(g)

route = geojson.LineString(c_acc)

with open('data/Route.geojson', 'w') as f:
    geojson.dump(route, f)
```
